```python
import pandas as pd
# from sklearn.model_selection import train_test_split

path = "/workspaces/A01_H07/data/melb_data.csv"
df = pd.read_csv(path)
```

```python
df.head()
```

```python
kullanilacak_sutunlar = ['Rooms', 'Distance', 'Landsize', 'BuildingArea', 'YearBuilt']
X = df[kullanilacak_sutunlar]

y = df['Price']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

my_pipeline = Pipeline(steps=[
    ('preprocessor', SimpleImputer()),
    ('model', RandomForestRegressor(n_estimators=50, random_state=0)),    
])

```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(estimator=my_pipeline, X=X, y=y, cv=5, scoring='neg_mean_absolute_error') * -1 # otomatik ypaiyor, evernotedaki resim durumunu cv 5 katmanda yap

print(scores)  # MAE scores   
print(scores.mean())  # MAE ortalama
```

## XG Boost Algoritmasi

```python
# Ayni data

kullanilacak_sutunlar = ['Rooms', 'Distance', 'Landsize', 'BuildingArea', 'YearBuilt']
X = df[kullanilacak_sutunlar]

y = df['Price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
!pip install xgboost
```

```python

from xgboost import XGBRegressor
```

```python
my_model = XGBRegressor()  # n_estimators parametresi
my_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
predictions = my_model.predict(X_test)
mae = mean_absolute_error(y_test, predictions)
print("MAE", mae)
```

```python
# n_estimators fazla verildiginde ezberleme olusuyor. Bunun onune gecmek icin optimizasyon yaparak ezberlemeyi cozebiliriz, loss grafikleri var "loss graphics xgboost" denildiginde
```

```python
my_model = XGBRegressor(n_estimators= 500, early_stopping_rounds=5)  # n_estimators parametresi
my_model.fit(X_train, y_train, eval_set=[(X_test, y_test)])
```

```python
# early_stopping_rounds =5 , eval_set komutlari
# calistiramadik python 3.10a geciyoruz, versiyon farki olabilir - calisti 3.10da

# bu early stopping rounds cok iyi bir optizasyon algoritmasi degil aslinda
```

```python
# Mola 14:20-14:56
```

```python
# zz: gelince, eval seti calistirdim oldu, eval set yazimi ile ilgili imis
```

```python
# early_stoppoing_round yerine, learning rate var, daha guzel 0.5 0.01 0.001 gibi kontrol noktalari sikilastirilabiliyor

my_model = XGBRegressor(n_estimators= 500, learning_rate=0.05)  # early stopping yerine LR tanimladik
my_model.fit(X_train, y_train, eval_set=[(X_test, y_test)])
```

```python
my_model = XGBRegressor(n_estimators= 500, learning_rate=0.05, early_stopping_rounds=5)  # LR'ye early stopping rounds eklersek
my_model.fit(X_train, y_train, eval_set=[(X_test, y_test)])
```

```python
# Paralel calisma vs sequential calisma

my_model = XGBRegressor(n_estimators= 500, learning_rate=0.05, early_stopping_rates=5, n_jobs=-1)  # 
my_model.fit(X_train, y_train, eval_set=[(X_test, y_test)])
```

```python
import pandas as pd

path = "/workspaces/A01_H07/data/AER_credit_card_data.csv"
df = pd.read_csv(path, true_values=['yes'], false_values=['no'])
```

```python
df
```

```python
X = df.drop(['card'], axis=1)
y = df['card']
```

```python
# veri sizintisi dogrudan overfitting verir - belki de overfitting engellenebiliyorsa leakage olmayabilir denemek lazim der hoca

from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier  # regressor degil classifier yes/no oldugu icin
from sklearn.model_selection import cross_val_score
```

```python
my_pipeline = make_pipeline(RandomForestClassifier(n_estimators=100))
cv_scores = cross_val_score(my_pipeline, X, y, cv=5, scoring='accuracy')

print(f"Cross Validation:\n{cv_scores}")
```

```python
# expenditure burada karti yoksa zaten 0 olacagi icin data leakage olarak sapittiriyor. %98 cok mukemmel
```

```python
kart_sahibi_harcamalari = X['expenditure'][y]  # kart sahibi olanlarin harcamalari
kart_sahibi_olmayanlarin_harcamalari = X['expenditure'][~y]

```

```python
kart_sahibi_olmayanlarin_harcamalari.sum()  
```

```python
potential_leaks = ['expenditure', 'active', 'majorcards', 'share']
X2 = X.drop(potential_leaks, axis=1)
```

```python
my_pipeline = make_pipeline(RandomForestClassifier(n_estimators=100))
cv_scores = cross_val_score(my_pipeline, X2, y, cv=5, scoring='accuracy')

print(f"Cross Validation:\n{cv_scores}")
```
